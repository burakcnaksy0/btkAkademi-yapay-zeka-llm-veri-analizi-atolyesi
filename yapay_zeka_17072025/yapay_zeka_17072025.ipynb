Eksik verilerle ***çalışmak***

isna() , fillna() , dropna() ile eksik veriler analiz edilir.
Eksik verileri silmek yerine , ortalama veya medyan gibi stratejilerle doldurmak daha etkilidir. Yeterli veri varsa eksik olan veri dışında kalanlar üzerinde tahmin yapıp eğer mevcut olanlar için geçerli oluyorsa eksik veriler içinde aynı işlem denenebilir.

```python
from ast import NotEq
import pandas as pd
df = pd.DataFrame({
    "urun" : ["kalem" , "defter" , "kalem" ,"silgi"],
    "adet": [10,None,8,3],
    "fiyat" : [2.5 ,5 ,None ,None]
})
print(df.isna())
print(df.isna().sum())
```

```python
average = df["adet"].mean()
df["adet"] = df["adet"].fillna(average)
print(f"Average adet : {average}")
print(df["adet"])
```

```python
average = df["fiyat"].mean()
df["fiyat"] = df["fiyat"].fillna(average)
print(f"Average price : {average}")
print(df["fiyat"])
```

SORU : En çok satış yapılan kategori neden bu kadar yüksek?
O kategoriye özel kampanya mı vardı? daha pahalı ürünler mi?

```python
import pandas as pd
df = pd.read_csv("satıslar.csv")
df["toplam"] = df["miktar"] * df["birim_fiyat"]

kategori_sat = df.groupby("kategori")["toplam"].sum()
print(kategori_sat)
```

**Soru**: en çok satış yapılan gün hangisi
**cevap** : idxmax() -> en yüksek değerin tarihi
**yorum** : hafta sonu mu ? kampanya dönemi mi ?

```python
import pandas as pd
df["tarih"] = pd.to_datetime(df["tarih"])
gunluk = df.groupby("tarih")["toplam"].sum()
print(gunluk.idxmax(), " gunu en cok satıs yapıldı")
```
