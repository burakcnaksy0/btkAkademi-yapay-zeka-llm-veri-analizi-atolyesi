```python
train = train.drop(["Employee ID","Employee Recognition", "Company Reputation","Performance Rating","Job Role"], axis=1)
test = test.drop(["Employee ID","Employee Recognition", "Company Reputation","Performance Rating","Job Role"], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train["Gender"] = le.fit_transform(train["Gender"])
test["Gender"] = le.fit_transform(test["Gender"])

train["Years at Company"] = le.fit_transform(train["Years at Company"])
test["Years at Company"] = le.fit_transform(test["Years at Company"])

train["Work-Life Balance"] = le.fit_transform(train["Work-Life Balance"])
test["Work-Life Balance"] = le.fit_transform(test["Work-Life Balance"])

train["Job Satisfaction"] = le.fit_transform(train["Job Satisfaction"])
test["Job Satisfaction"] = le.fit_transform(test["Job Satisfaction"])

train["Education Level"] = le.fit_transform(train["Education Level"])
test["Education Level"] = le.fit_transform(test["Education Level"])

train["Company Size"] = le.fit_transform(train["Company Size"])
test["Company Size"] = le.fit_transform(test["Company Size"])

train["Overtime"] = le.fit_transform(train["Overtime"])
test["Overtime"] = le.fit_transform(test["Overtime"])

train["Attrition"] = le.fit_transform(train["Attrition"])
test["Attrition"] = le.fit_transform(test["Attrition"])

train["Job Level"] = le.fit_transform(train["Job Level"])
test["Job Level"] = le.fit_transform(test["Job Level"])

train["Marital Status"] = le.fit_transform(train["Marital Status"])
test["Marital Status"] = le.fit_transform(test["Marital Status"])

train["Remote Work"] = le.fit_transform(train["Remote Work"])
test["Remote Work"] = le.fit_transform(test["Remote Work"])

train["Leadership Opportunities"] = le.fit_transform(train["Leadership Opportunities"])
test["Leadership Opportunities"] = le.fit_transform(test["Leadership Opportunities"])

train["Innovation Opportunities"] = le.fit_transform(train["Innovation Opportunities"])
test["Innovation Opportunities"] = le.fit_transform(test["Innovation Opportunities"])
```

```python
train
```

```python
test
```

```python
X_train , y_train = train.drop(["Attrition"], axis=1), train["Attrition"]
```

```python
X_test , y_test = test.drop(["Attrition"], axis=1), test["Attrition"]
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler


y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , InputLayer
from tensorflow.keras.regularizers import l2

model = Sequential([
    InputLayer(input_shape=(19,)),
    Dense(128,activation='relu',kernel_regularizer=l2(0.01)),
    Dropout(0.5), #yüzde 50 oranında overfitting engellemek için kullanılır.
    Dense(128,activation='relu',kernel_regularizer=l2(0.01)),
    Dropout(0.5),
    Dense(128,activation='relu',kernel_regularizer=l2(0.01)),
    Dropout(0.5),
    Dense(2,activation="softmax")
])
```

```python
model.compile(
    optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3)

history = model.fit(
    X_train,y_train,
    validation_split = 0.2,
    epochs=50,
    batch_size=32,
    verbose = 1,
    callbacks=[early_stopping]
)
```

```python
test_loss,test_acc = model.evaluate(X_test,y_test)
print(f"Test Accuracy: {test_acc}")

sample = X_test[0].reshape(1,-1)
prediction = model.predict(sample)
print(f"Prediction: {prediction}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Accuracy: {accuracy}")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm_model = SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"SVM Accuracy: {accuracy}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN Accuracy: {accuracy}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred = dt_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision Tree Accuracy: {accuracy}")
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred = nb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Naive Bayes Accuracy: {accuracy}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
y_pred = lr_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Logistic Regression Accuracy: {accuracy}")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp_model.fit(X_train, y_train)
y_pred = mlp_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"MLP Accuracy: {accuracy}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
y_pred = gb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Gradient Boosting Accuracy: {accuracy}")
```
